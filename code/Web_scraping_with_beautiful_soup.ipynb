```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://dl.acm.org/doi/proceedings/10.1145/2814864"
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
soup.find_all("a")
```

```python
for link in soup.find_all("a"):
    print(link)
    print("-------")
```

```python
soup.title.string
```

```python
soup.title.parent.name
```

```python
for publication in soup.find_all("h5"):
    print(publication)
    print("---------------")
```

```python
for publication in soup.find_all("h5", {"class": "issue-item__title"}):
    print(publication)
    print("---------------")
```

```python
for publication in soup.find_all("h5", {"class": "issue-item__title"}):
    print(f"Title: {publication.string}")
    print(f'DOI: {publication.a["href"].replace("/doi/", "")}')
    print("---------------")
```

```python
with open("title_and_doi.tsv", "w") as title_fh:
    title_fh.write("title\tdoi\n")
    for publication in soup.find_all("h5", {"class": "issue-item__title"}):
        # print(f"Title: {publication.string}")
        # print(f'DOI: {publication.a["href"].replace("/doi/", "")}')
        # print("---------------")
        title_fh.write(
            f'{publication.string}\t{publication.a["href"].replace("/doi/", "")}\n')
```
